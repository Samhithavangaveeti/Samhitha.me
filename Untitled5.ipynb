MATPLOTLIB

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 1. Line Plot (Trends over time)
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure(figsize=(6, 4))
plt.plot(x, y, label='sin(x)', color='b')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot of sin(x)')
plt.legend()
plt.grid(True)
plt.show()

# 2. Bar Chart (Comparing categories)
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
plt.bar(categories, values, color=['red', 'blue', 'green', 'orange'])
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()

# 3. Scatter Plot (Relationship between variables)
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y, color='purple', alpha=0.6)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()

# 4. Histogram (Distribution of data)
data = np.random.randn(1000)
plt.hist(data, bins=30, color='green', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()

# 5. Pie Chart (Proportions of categories)
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [20, 30, 25, 25]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'yellow', 'red', 'brown'])
plt.title('Pie Chart')
plt.show()

# 6. Box Plot (Identifying outliers)
data = [np.random.randn(100), np.random.randn(100) + 1]
plt.boxplot(data, labels=['Group 1', 'Group 2'])
plt.title('Box Plot')
plt.show()

# 7. Violin Plot (Density + Box Plot)
plt.violinplot([np.random.randn(100), np.random.randn(100) + 1])
plt.title('Violin Plot')
plt.show()

# 8. Heatmap using Matplotlib
matrix = np.random.rand(5, 5)
plt.imshow(matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.title('Heatmap')
plt.show()

# 9. 3D Surface Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('3D Surface Plot')
plt.show()

# 10. Time Series Plot
t = np.arange(0., 5., 0.2)
plt.plot(t, t**2, 'r--', label='t^2')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time Series Plot')
plt.legend()
plt.show()
```

SEABORN

```python
import seaborn as sns
import pandas as pd

# Sample Data
df = sns.load_dataset('tips')

# 1. Pair Plot (Pairwise relationships)
sns.pairplot(df, hue='sex')
plt.show()

# 2. Swarm Plot (Categorical scatter)
sns.swarmplot(x='day', y='total_bill', data=df)
plt.title('Swarm Plot')
plt.show()

# 3. Violin Plot (Density + Box)
sns.violinplot(x='day', y='total_bill', data=df, hue='sex', split=True)
plt.title('Violin Plot')
plt.show()

# 4. Heatmap (Correlation Matrix)
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()

# 5. Box Plot (Summary Statistics)
sns.boxplot(x='day', y='total_bill', data=df)
plt.title('Box Plot')
plt.show()

# 6. Bar Plot (Category comparison)
sns.barplot(x='day', y='total_bill', data=df, estimator=np.mean)
plt.title('Bar Plot')
plt.show()

# 7. Count Plot (Category frequency)
sns.countplot(x='day', data=df)
plt.title('Count Plot')
plt.show()

# 8. Scatter Plot with Regression (Regression trends)
sns.regplot(x='total_bill', y='tip', data=df)
plt.title('Regression Plot')
plt.show()

# 9. KDE Plot (Density Estimation)
sns.kdeplot(df['total_bill'], shade=True)
plt.title('KDE Plot')
plt.show()

# 10. Facet Grid (Multiple plots in one)
g = sns.FacetGrid(df, col='sex')
g.map(plt.hist, 'total_bill')
plt.show()
```
